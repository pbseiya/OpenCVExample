```python
import matplotlib.pyplot as plt
from PIL import Image
import cv2
```

```python
im = cv2.imread('parrot.jpg')
plt.imshow(im[:,:,::-1])
```

```python
im = plt.imread('parrot.jpg')
plt.imshow(im)
```

```python
fig, (ax1, ax2, ax3) = plt.subplots(1, 3,figsize=(14, 6))
fig.suptitle('Horizontally stacked subplots')
ax1.imshow(im[:,:,2])
ax2.imshow(im[:,:,1])
ax3.imshow(im[:,:,0])
```
