```python
# Import Necessary Libraries
import cv2
import numpy as np
import matplotlib.pyplot as plt
```

```python
# Draw a grayscale (Number of channel = 1) Square
square_image = np.zeros((400,400,1),np.uint8)
cv2.rectangle(square_image,(75,75), (300,300), 255, -1)
plt.figure(figsize=(30,20))
plt.imshow(square_image,cmap='gray');
```

```python
# Draw greyscale (Number of channel = 1) Circle
circle = np.zeros((400,400,1), np.uint8)
cv2.circle(circle, (300,300), 75, 255, -1)
plt.figure(figsize=(30,20))
plt.imshow(circle, cmap='gray');
```

# Bitwise Operation on Shapes

```python
# Showing the region where both square and circle intersections
AND_Operation = cv2.bitwise_and(square_image, circle)
plt.figure(figsize=(30,20))
plt.title('AND_Operation')
plt.imshow(AND_Operation);
```

```python
# Showing the region where either square or circle is
OR_Operation = cv2.bitwise_or(square_image, circle)
plt.figure(figsize=(30,20))
plt.title('OR_Operation')
plt.imshow(OR_Operation);
```

```python
# Showing the region where either square and circle exits
XOR_Operation = cv2.bitwise_xor(square_image, circle)
plt.figure(figsize=(30,20))
plt.title('XOR_Operation')
plt.imshow(XOR_Operation);
```

```python
# Showing the region that isn't part of the square i.e. inverse of the square image
NOT_Operation = cv2.bitwise_not(square_image)
plt.figure(figsize=(30,20))
plt.title('NOT_Operation')
plt.imshow(NOT_Operation);
```

```python
# Showing the region where either square and circle exits
XOR_Operation = cv2.bitwise_xor(square_image, circle)
plt.figure(figsize=(30,20))
plt.title('XOR_Operation')
plt.imshow(XOR_Operation);
```
